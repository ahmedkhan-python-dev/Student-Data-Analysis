### **Central Tendency**

**Mean**                                      
The formula of finding mean is **sum of data / len of data**

```python
age_mean = df["Age"].mean()
score_mean = df["Score"].mean()
attend_mean = df["Attendance (%)"].mean()
print({"Average Age": age_mean, "Average Score": score_mean, "Average Attendance": attend_mean})
```

**Median**                                
The formula or way of finding median is **if data have odd number of data than middle value is 1 and our answer elif data have even number of data than middle value is 2 and sum of these two values and divided by 2 the answer we called Median**

```python
age_median = df["Age"].median()
score_median = df["Score"].median()
attend_median = df["Attendance (%)"].median()
print({"Median Age": age_median, "Median Score": score_median, "Median Attendance": attend_median})
```

**Mode**                 
The formula of finding the mode is **most repeating value or number in dataset is Mode**

```python
age_mode = df["Age"].mode()[0]
score_mode = df["Score"].mode()[0]
attend_mode = df["Attendance (%)"].mode()[0]
print({"Age Mode is": age_mode, "Score Mode is": score_mode, "Attendance Mode is": attend_mode})
```

### **Variability**

**Range**         
The formula of finding the range is **Max value - Min value of dataset**                                        
***Caution:*** It can be wrong because outliers avaibility in dataset

**Mean Absolute Deviation**                  
The formula is to find MTD is **Sum( | dataset - mean | ) / len of dataset**

```python
age_mtd = np.sum(np.abs(df["Age"] - age_mean)) / len(df["Age"])
score_mtd = np.sum(np.abs(df["Score"] - score_mean)) / len(df["Score"])
attend_mtd = np.sum(np.abs(df["Attendance (%)"] - attend_mean)) / len(df["Attendance (%)"])
print({"Age MTD": age_mtd, "Score MTD": score_mtd, "Attendance MTD": attend_mtd})
```

**Standard Deviation**             
The formula is to find STD is **Sqaure Root ( Sum (dataset - mean) ** 2) len(dataset)**

```python
age_std = np.sqrt(np.sum((df["Age"] - age_mean) ** 2) / len(df["Age"]))
score_std = np.sqrt(np.sum((df["Score"] - score_mean) ** 2) / len(df["Score"]))
attend_std = np.sqrt(np.sum((df["Attendance (%)"] - attend_mean) ** 2) / len(df["Attendance (%)"]))
print({"Age STD": age_std, "Score STD": score_std, "Attendance STD": attend_std})
```

**Variance**            
The formula is to find variance is **Standard Deviation ** 2**

```python
age_var = age_std ** 2
score_var = score_std ** 2
attend_var = attend_std ** 2
print({"Age Variance": age_var, "Score Variance": score_var, "Attendance Variance": attend_var})
```

### **Hypothesis Testing**

Here I'll apply **T - test** method

```python
male_students = df[df["Gender"] == "Male"]
female_students = df[df["Gender"] == "Female"]
```

```python
samp_male = male_students.sample(10)
samp_female = female_students.sample(10)
```

```python
samp_male.head()
```

```python
samp_female.head()
```

```python
samp_male_mean = samp_male["Score"].mean()
samp_female_mean = samp_female["Score"].mean()
print({"Male Students Score Average": samp_male_mean, "Female Students Score Average": samp_female_mean})
```

```python
samp_male_std = samp_male["Score"].std()
samp_female_std = samp_female["Score"].std()
print({"Male Students Score STD": samp_male_std, "Female Students Score STD": samp_female_std})
```

```python
samp_quantity = 10
alpha = 0.025 # because 5% = 0.05 and 0.05 / 2 = 0.025
```

Applying Formula

```python
ttest_cal_numerator = samp_male_mean - samp_female_mean
ttest_cal_denomirator = np.sqrt(((samp_male_std ** 2) / samp_quantity) + ((samp_female_std ** 2) / samp_quantity))
ttest_cal = ttest_cal_numerator / ttest_cal_denomirator
ttest = st.t.ppf(1 - alpha, samp_quantity + samp_quantity - 2)
```

```python
print({"T - test Calculation": ttest_cal, "T - test Value": ttest})
```

```python
if ttest_cal == ttest:
  print("H0 is right. Both are equal")
else:
  print("H0 is wrong. Both are not equal")
```
